```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, classification_report, confusion_matrix

# Step 2: Load dataset
df = pd.read_csv("/content/Synthetic_Gaming_Data.csv")  # update if needed
df.columns = df.columns.str.strip().str.replace(r'\s+', ' ', regex=True)
df = df.drop_duplicates()

# Step 3: Convert Yes/No to binary
yes_no_cols = [
    'Do you often lose track of time while gaming?',
    'Have you ever skipped meals or sleep due to gaming?',
    'Have others ever expressed concern about your gaming habits?',
    'Have you ever felt guilty or depressed after long gaming sessions?',
    'Do you think gaming helps you cope with stress or emotional issues?',
    'Do you consent to your responses being used anonymously for academic research?'
]
for col in yes_no_cols:
    df[col] = df[col].astype(str).str.strip().str.lower().map({'yes': 1, 'no': 0})

# Step 4: Map social withdrawal levels
withdrawal_map = {'Never': 0, 'Rarely': 1, 'Sometimes': 2, 'Often': 3}
df['How often do you feel socially withdrawn or isolated?'] = (
    df['How often do you feel socially withdrawn or isolated?']
    .astype(str).str.strip().map(withdrawal_map)
)

# Step 5: Feature Engineering
df['gaming_intensity'] = df['How many hours do you play video games per day on average?'] * df['How many days per week do you play games?']
df['behavioral_flags'] = (
    df['Do you often lose track of time while gaming?'] +
    df['Have you ever skipped meals or sleep due to gaming?'] +
    df['Have others ever expressed concern about your gaming habits?'] +
    df['Have you ever felt guilty or depressed after long gaming sessions?']
)
df['mental_burden'] = df['On a scale of 1 to 10, how stressed do you feel on average?'] + df['On a scale of 1 to 10, how anxious do you feel regularly?']
df['sleep_deprived'] = (df['On average, how many hours of sleep do you get per night?'] < 5).astype(int)

# Step 6: Create risk score and risk level
def calculate_risk(row):
    score = 0
    if row['How many hours do you play video games per day on average?'] >= 5:
        score += 2
    if row['How many days per week do you play games?'] >= 5:
        score += 2
    if row['Have you ever skipped meals or sleep due to gaming?'] == 1:
        score += 2
    if row['Have you ever felt guilty or depressed after long gaming sessions?'] == 1:
        score += 2
    if row['Do you often lose track of time while gaming?'] == 1:
        score += 2
    return score

df['Gaming_Risk_Score'] = df.apply(calculate_risk, axis=1)

def risk_level(score):
    if score <= 3:
        return 'Low'
    elif score <= 6:
        return 'Medium'
    else:
        return 'High'

df['Gaming_Risk_Level'] = df['Gaming_Risk_Score'].apply(risk_level)

# Step 7: Encode categorical columns
label_encoders = {}
cat_cols = df.select_dtypes(include='object').columns.drop('Gaming_Risk_Level', errors='ignore')
for col in cat_cols:
    df[col] = df[col].astype(str)
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# Step 8: Prepare model data
X = df.drop(columns=['Gaming_Risk_Score', 'Gaming_Risk_Level'])
y = df['Gaming_Risk_Score']

# Normalize features
X = X.apply(pd.to_numeric, errors='coerce').dropna(axis=1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 9: Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Step 10: Train regression model
model = RandomForestRegressor(random_state=42)
model.fit(X_train, y_train)

# Step 11: Predict and evaluate regression
y_train_pred = model.predict(X_train)
y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("Regression Evaluation:")
print("MAE:", mae)
print("RMSE:", rmse)
print("R:", r2)

# Step 12: Convert predicted scores to risk levels
def score_to_level(score):
    if score <= 3:
        return 'Low'
    elif score <= 6:
        return 'Medium'
    else:
        return 'High'

y_pred_levels = pd.Series(y_pred).apply(score_to_level)
y_true_levels = y_test.apply(score_to_level)

# Step 13: Evaluate classification quality
print("\nClassification Evaluation (Based on Predicted Scores):")
print(classification_report(y_true_levels, y_pred_levels))

conf_matrix = confusion_matrix(y_true_levels, y_pred_levels, labels=["Low", "Medium", "High"])
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=["Low", "Medium", "High"],
            yticklabels=["Low", "Medium", "High"])
plt.title("Confusion Matrix (Levels from Predicted Scores)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Step 14: Overfitting check (train vs test)
train_r2 = r2_score(y_train, y_train_pred)
print(f"\nOverfitting Check:\nTrain R: {train_r2:.3f} | Test R: {r2:.3f}")
```
